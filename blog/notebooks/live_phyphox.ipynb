# Example on how to get the meausrements from phone in real time

## Check the available signals

```python
import subprocess, json, time

base_url = "http://192.168.1.11"  # change if your phone has a different IP

# Read config from phone and print available / recommended buffers
def get_config():
    r = subprocess.run(["curl", "-s", f"{base_url}/config"], capture_output=True, text=True)
    return json.loads(r.stdout)

cfg = get_config()
print("Buffers available:", [b.get('name') for b in cfg.get('buffers', [])])
print("Recommended exports:", [e.get('buffer') for e in cfg.get('export', [])])
```

## Get the data in real time

```python
import subprocess, json, time

url = "http://192.168.1.11/get?accY"

def fetch_json():
    result = subprocess.run(["curl", "-s", url], capture_output=True, text=True)
    return json.loads(result.stdout)

# Loop at ~100 Hz
while True:
    data = fetch_json()
    raw_value = data["buffer"]["accY"]["buffer"][0]

    if raw_value is None:
        print("No data yetâ€¦")
    else:
        print(f"accY = {raw_value:.6f}")

    time.sleep(0.01)  # 100 Hz
```
